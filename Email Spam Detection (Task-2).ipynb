```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, f1_score, confusion_matrix

def plot_confusion_matrix(y_true, y_pred, labels):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()

def evaluate_naive_bayes_model(X_train, Y_train, X_test, Y_test):
    # Create and train the Naive Bayes model
    model_nb = GaussianNB()
    model_nb.fit(X_train, Y_train)
    
    # Make predictions on the test set
    y_pred_nb = model_nb.predict(X_test)
    
    # Calculate metrics
    accuracy = accuracy_score(Y_test, y_pred_nb)
    precision = precision_score(Y_test, y_pred_nb)
    f1 = f1_score(Y_test, y_pred_nb)

    print("Accuracy Score:", accuracy)
    print("Precision:", precision)
    print("F1 Score:", f1)
    print("")

    # Plot the confusion matrix using seaborn
    display_labels = ['Spam', 'Ham']
    plot_confusion_matrix(Y_test, y_pred_nb, labels=display_labels)

    return {
        "accuracy": accuracy,
        "precision": precision,
        "f1_score": f1,
    }

# Example usage:
metrics_result = evaluate_naive_bayes_model(X_train, Y_train, X_test, Y_test)
```
